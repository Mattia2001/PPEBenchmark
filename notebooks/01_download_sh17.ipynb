```python
from pathlib import Path

repo_dir = SH17_DIR / "sh17dataset"

for item in repo_dir.iterdir():
    print(item.name)
```

```python
import os

def get_dir_size(path: Path) -> float:
    total = 0
    for root, dirs, files in os.walk(path):
        for file in files:
            total += os.path.getsize(os.path.join(root, file))
    return total / (1024 ** 3)

print(f"SH17 repository size: {get_dir_size(repo_dir):.2f} GB")
```
